```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./Student_Performance (1).csv")

```

```python
df['Extracurricular Activities']  = df['Extracurricular Activities'].apply(lambda x : 1 if x == 'Yes' else 0)
df
```

```python
sns.scatterplot(data= df , x= 'Hours Studied' , y='Performance Index' , alpha = 0.5 , hue = 'Performance Index');
```

```python
sns.relplot(data= df , x= 'Previous Scores' , y='Performance Index' , hue = 'Performance Index');
```

```python

sns.stripplot(data= df.sample(100) , x = 'Sleep Hours' , y = "Performance Index" )
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train, x_test  , y_train , y_test = train_test_split(df.drop(columns=["Performance Index"]) , df['Performance Index'])
```

```python
lm_model = LinearRegression()
lm_model.fit(x_train , y_train)

mean_absolute_error(y_test , lm_model.predict(x_test))
```

```python
import joblib
joblib.dump(lm_model , "Linear_reg_model.joblib")
```
